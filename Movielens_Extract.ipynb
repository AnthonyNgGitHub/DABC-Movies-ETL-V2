### Initial Look at the Movie Metadata

Check that all of the columns came in as the correct data types.

A few columns that need to be converted:

1. video should be Boolean

2. adult should be Boolean

3. popularity should be numeric

##### Before we convert the "adult" and "video" columns, we want to check that all the values are either True or False.

```python
kaggle_metadata['adult'].value_counts()
```

Remove the bad data

- Recall : putting the "~" infront of a boolean assessment will give the opposite results

```python
kaggle_metadata[~kaggle_metadata['adult'].isin(['True','False'])]
```

The biggest concern is that none of the data in these rows looks like an imdb_id.

Since that's missing, there's no amount of rearranging that will make these rows into good data.

We're just going to have to drop them.

In fact, since we probably don't want to include adult movies in the hackathon dataset, we'll only keep rows where adult is False, and then drop the "adult" column.

```python
# Keep rows where the adult column is False, and then drop the adult column.
kaggle_metadata = kaggle_metadata[kaggle_metadata['adult'] == 'False'].drop('adult',axis='columns')
```

##### Video column

```python
kaggle_metadata['video'].value_counts()
```

```python
# Create boolean column
kaggle_metadata['video'] == 'True'
```

```python
# Assign boolean column back to video
kaggle_metadata['video'] = kaggle_metadata['video'] == 'True'
```

##### Numeric columns

Use the to_numeric() method from Pandas.

We'll make sure the 

    errors= 
    
argument is set to 'raise', so we'll know if there's any data that can't be converted to numbers.

```python
kaggle_metadata['budget'] = kaggle_metadata['budget'].astype(int)
kaggle_metadata['id'] = pd.to_numeric(kaggle_metadata['id'], errors='raise')
kaggle_metadata['popularity'] = pd.to_numeric(kaggle_metadata['popularity'], errors='raise')
```

##### Convert release_date to datetime

Luckily, release_date is in a standard format.

    to_datetime()
    
  will convert it without any fuss.

```python
kaggle_metadata['release_date'] = pd.to_datetime(kaggle_metadata['release_date'])
```

### Reasonability Checks on Ratings Data

```python
ratings.info(null_counts=True)
```

For our own analysis, we won't be using the timestamp column;

However, we will be storing the rating data as its own table in SQL, so we'll need to convert it to a datetime data type. 

From the MovieLens documentation, the timestamp is the number of seconds since midnight of January 1, 1970.

##### Note

Storing time values as a data type is difficult, and there are many, many standards out there for time values.

- Some store time values as text strings, like the ISO format "1955-11-05T12:00:00," but then calculating the difference between two time values is complicated and computationally expensive.


- The Unix time standard stores points of time as integers, specifically as the number of seconds that have elapsed since midnight of January 1, 1970. This is known as the Unix epoch. 


- There are other epochs in use, but the Unix epoch is by far the most widespread.

We'll specify in 
    
    to_datetime()
   
   that the origin is 'unix' and the time unit is seconds.

```python
pd.to_datetime(ratings['timestamp'], unit='s')
```

```python
# Asign it to the timestamp column
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')
```
